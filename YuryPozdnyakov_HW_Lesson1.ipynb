__Задача 1__

Группа студентов изучает 10 различных дисциплин. Сколькими способами можно составить расписание на понедельник, если в этот день должно быть 4 разных занятия?

Воспользуемся формулой комбинаторики, чтобы вычислить количество сочетаний:

### $$C_n^k = \frac{n!}{k!(n - k)!}$$

```python
from math import factorial

def combinations(n, k):
    return int(factorial(n) / (factorial(k) * factorial(n - k)))
```

```python
combinations(10, 4)
```

__Задача 2__

Из колоды в 52 карты вынимают случайным образом 4 карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

__Задача 3__

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что два определённых человека будут сидеть рядом?

```python
def permutations(n):
    return int(factorial(n))
```

 Позиция пары людей в ряду описывается выражением n-1, таким образом есть 6 разных позиций. Относительно друг друга люди в паре могут распологаться двумя способами. Остальные 5 человек могут распологаться 120 способами

```python
p5 = permutations(5)
p5
```

Из этого следует, что количество благоприятных исходов будет равно:

```python
m = 6 * 2 * p5
m
```

Всего исходов:

```python
n = permutations(7)
n
```

Отсюда следует, что вероятность того, что два определённых человека будут сидеть рядом вычисляется по формуле:

### $$P(A) = \frac{m}{n}$$

```python
p = m / n
p
```

__а)__

Всего исходов:

```python
n = combinations(60,3)
n
```

Благоприятных исходов:

```python
m = combinations(50,3)
m
```

Вычислим вероятность выбора трех вопросов, которые знает студент:

```python
p = m / n
p
```

__б)__

Число исходов, когда выбираются 2 выученных вопроса:

```python
m2from50 = combinations(50,2)
m2from50
```

Число невыученных вопросов 60 - 50 = 10

Число исходов, когда выбираются 1 невыученный вопрос:

```python
m1from10 = combinations(10,1)
m1from10
```

Всего благоприятных исходов:

```python
m = m2from50 * m1from10
m
```

Вероятность выбора двух выученных вопросов из тройки вытянутых студентом:

```python
p = m / n
p
```
