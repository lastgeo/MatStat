Вероятность случайного угадывания правильного ответа в отдельно взятом вопросе равна:

### $$p=\frac{m}{n}=\frac{1}{4}=0.25$$

Посчитаем распределение числа угадывания правильного ответа при пяти вопросах в контрольной работе. 


```python
import numpy as np
```

```python
def combinations(n, k):
    return int(np.math.factorial(n) / (np.math.factorial(k) * np.math.factorial(n - k)))

def binominal(n, k, p):
    return combinations(n, k) * (p ** k) * ((1 - p) ** (n - k))

# Количество вопросов
n = 5

# Вероятность угадывания правильного ответа из четырех
p1 = 1 / 4
```

```python
# Закон распределения
p = []

for k in range(n + 1):
    p.append(binominal(n, k, p1))
p
```

__Задача 2__

Вероятность поймать рыбу при одном забрасывании удочки равна 0.01. Какова вероятность того, что число пойманных рыб при 200 забросах удочки будет не менее 5 и не более 10?

$a = 200 \cdot 0.01 = 2$

$k\in[5,10]$

```python
def pois(a, k):
    return (a ** k) * (np.exp(-a)) / np.math.factorial(k)
```

```python
a = 2

# Сумма вероятностей того, что число пойманных рыб 
# будет не менее 5 и не более 10
sum(pois(a=a,k=i) for i in range(5,11))
```

__Задача 3__

В условиях предыдущей задачи сколько забросов удочки нужно сделать чтобы быть уверенным с вероятностью 0.9, что хотя бы 10 раз будет поймана рыба?

_Подсказка_. Как посчитать `P(X >= k)` для величины `X`, имеющей биномиальное распределение? Здесь предстоит немножко покодить.

```python
# Требуемая вероятность
p_u = 0.9
# Требуемое число пойманных рыб
k = 10
# Вероятность поймать рыбу
p_n = 0.01
# Данные для инициализаци
p = 1
n = 0

while p > (1 - p_u):
    n += 1
    a = n * p_n
    p = sum(pois(a=a,k=i) for i in range(k))

print(n,(1 - p))
```
