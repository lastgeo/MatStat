## –¢–µ–º–∞ ¬´–î–∏—Å–∫—Ä–µ—Ç–Ω—ã–µ —Å–ª—É—á–∞–π–Ω—ã–µ –≤–µ–ª–∏—á–∏–Ω—ã. –ó–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π. –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–π –∑–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ü—É–∞—Å—Å–æ–Ω–∞¬ª

__–ó–∞–¥–∞—á–∞ 1__

–ö–æ–Ω—Ç—Ä–æ–ª—å–Ω–∞—è —Ä–∞–±–æ—Ç–∞ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –ø—è—Ç–∏ –≤–æ–ø—Ä–æ—Å–æ–≤. –ù–∞ –∫–∞–∂–¥—ã–π –≤–æ–ø—Ä–æ—Å –ø—Ä–∏–≤–µ–¥–µ–Ω–æ —á–µ—Ç—ã—Ä–µ –≤–∞—Ä–∏–∞–Ω—Ç–∞ –æ—Ç–≤–µ—Ç–∞, –æ–¥–∏–Ω –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–∞–≤–∏–ª—å–Ω—ã–π. –°–æ—Å—Ç–∞–≤—å—Ç–µ –∑–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —á–∏—Å–ª–∞ –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –ø—Ä–∏ –ø—Ä–æ—Å—Ç–æ–º —É–≥–∞–¥—ã–≤–∞–Ω–∏–∏. –ù–∞–π–¥–∏—Ç–µ `M(X), D(X)`.

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ —É–≥–∞–¥—ã–≤–∞–Ω–∏—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –æ—Ç–≤–µ—Ç–∞ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ –≤–∑—è—Ç–æ–º –≤–æ–ø—Ä–æ—Å–µ —Ä–∞–≤–Ω–∞:

### $$p=\frac{m}{n}=\frac{1}{4}=0.25$$

–ü–æ—Å—á–∏—Ç–∞–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —á–∏—Å–ª–∞ —É–≥–∞–¥—ã–≤–∞–Ω–∏—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –æ—Ç–≤–µ—Ç–∞ –ø—Ä–∏ –ø—è—Ç–∏ –≤–æ–ø—Ä–æ—Å–∞—Ö –≤ –∫–æ–Ω—Ç—Ä–æ–ª—å–Ω–æ–π —Ä–∞–±–æ—Ç–µ. –í–æ–∑–º–æ–∂–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–∞–∫–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã:  

$ùë•_1=0,  ùë•_2=1,  ùë•_3=2,  ùë•_4=3, ùë•_5=4, ùë•_6=5.$

–ü–æ—Å—á–∏—Ç–∞–µ–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è:

```python
import numpy as np
```

```python
def combinations(n, k):
    return int(np.math.factorial(n) / (np.math.factorial(k) * np.math.factorial(n - k)))

def binominal(n, k, p):
    return combinations(n, k) * (p ** k) * ((1 - p) ** (n - k))

# –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤–æ–ø—Ä–æ—Å–æ–≤
n = 5

# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É–≥–∞–¥—ã–≤–∞–Ω–∏—è –ø—Ä–∞–≤–∏–ª—å–Ω–æ–≥–æ –æ—Ç–≤–µ—Ç–∞ –∏–∑ —á–µ—Ç—ã—Ä–µ—Ö
p1 = 1 / 4
```

```python
# –ó–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
p = []

for k in range(n + 1):
    p.append(binominal(n, k, p1))
p
```

–ü–æ—Å—á–∏—Ç–∞–µ–º –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –ü–æ—Å–∫–æ–ª—å–∫—É $n = 5$ –∏ $p = 0.25$, –ø–æ–ª—É—á–∞–µ–º:

$$M(X) = 5 \cdot 0.25 = 1.25$$

–î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞–∫–æ–π –≤–µ–ª–∏—á–∏–Ω—ã:

$$D(X) = 5 \cdot 0.25 \cdot (1 - 0.25) = 0.9375$$

__–ó–∞–¥–∞—á–∞ 2__

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–π–º–∞—Ç—å —Ä—ã–±—É –ø—Ä–∏ –æ–¥–Ω–æ–º –∑–∞–±—Ä–∞—Å—ã–≤–∞–Ω–∏–∏ —É–¥–æ—á–∫–∏ —Ä–∞–≤–Ω–∞ 0.01. –ö–∞–∫–æ–≤–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —á–∏—Å–ª–æ –ø–æ–π–º–∞–Ω–Ω—ã—Ö —Ä—ã–± –ø—Ä–∏ 200 –∑–∞–±—Ä–æ—Å–∞—Ö —É–¥–æ—á–∫–∏ –±—É–¥–µ—Ç –Ω–µ –º–µ–Ω–µ–µ 5 –∏ –Ω–µ –±–æ–ª–µ–µ 10?

$a = 200 \cdot 0.01 = 2$

$k\in[5,10]$

```python
def pois(a, k):
    return (a ** k) * (np.exp(-a)) / np.math.factorial(k)
```

```python
a = 2

# –°—É–º–º–∞ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π —Ç–æ–≥–æ, —á—Ç–æ —á–∏—Å–ª–æ –ø–æ–π–º–∞–Ω–Ω—ã—Ö —Ä—ã–± 
# –±—É–¥–µ—Ç –Ω–µ –º–µ–Ω–µ–µ 5 –∏ –Ω–µ –±–æ–ª–µ–µ 10
sum(pois(a=a,k=i) for i in range(5,11))
```

__–ó–∞–¥–∞—á–∞ 3__

–í —É—Å–ª–æ–≤–∏—è—Ö –ø—Ä–µ–¥—ã–¥—É—â–µ–π –∑–∞–¥–∞—á–∏ —Å–∫–æ–ª—å–∫–æ –∑–∞–±—Ä–æ—Å–æ–≤ —É–¥–æ—á–∫–∏ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —á—Ç–æ–±—ã –±—ã—Ç—å —É–≤–µ—Ä–µ–Ω–Ω—ã–º —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 0.9, —á—Ç–æ —Ö–æ—Ç—è –±—ã 10 —Ä–∞–∑ –±—É–¥–µ—Ç –ø–æ–π–º–∞–Ω–∞ —Ä—ã–±–∞?

_–ü–æ–¥—Å–∫–∞–∑–∫–∞_. –ö–∞–∫ –ø–æ—Å—á–∏—Ç–∞—Ç—å `P(X >= k)` –¥–ª—è –≤–µ–ª–∏—á–∏–Ω—ã `X`, –∏–º–µ—é—â–µ–π –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ? –ó–¥–µ—Å—å –ø—Ä–µ–¥—Å—Ç–æ–∏—Ç –Ω–µ–º–Ω–æ–∂–∫–æ –ø–æ–∫–æ–¥–∏—Ç—å.

```python
# –¢—Ä–µ–±—É–µ–º–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
p_u = 0.9
# –¢—Ä–µ–±—É–µ–º–æ–µ —á–∏—Å–ª–æ –ø–æ–π–º–∞–Ω–Ω—ã—Ö —Ä—ã–±
k = 10
# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–π–º–∞—Ç—å —Ä—ã–±—É
p_n = 0.01
# –î–∞–Ω–Ω—ã–µ –¥–ª—è –∏–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏
p = 1
n = 0

while p > (1 - p_u):
    n += 1
    a = n * p_n
    p = sum(pois(a=a,k=i) for i in range(k))

print(n,(1 - p))
```
