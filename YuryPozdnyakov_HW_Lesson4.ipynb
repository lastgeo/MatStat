__Задача 1__

О непрерывной равномерно распределенной случайной величине `B` известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины `B` и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

Дисперсию $D(X)$ равномерно распределенной непрерывной случайной величины можно вычислить по формуле:

$$D(X) = \frac{(b - a)^2}{12},$$

из этого следует, что правая граница находится по формуле:

$$b = \sqrt{12D(X)}+a$$

а среднее значение равно математическому ожиданию и вычисляется по формуле:

$$M(X) = \frac{a + b}{2}$$

```python
d = 0.2
a = 0.5
b = (12 * d) ** 0.5 + a
# правая граница
b
```

```python
m = (a + b) / 2
# среднее значение
m
```
