Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. Даны значения роста в трех группах случайно выбранных спортсменов: 
```
football_players = [173, 175, 180, 178, 177, 185, 183, 182]
hockey_players = [177, 179, 180, 188, 177, 172, 171, 184, 180]
lifters = [172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170]
```

````python
import numpy as np

football_players = np.array([173, 175, 180, 178, 177, 185, 183, 182])
hockey_players = np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])
lifters = np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])

football_players_n = football_players.shape[0]
hockey_players_n = hockey_players.shape[0]
lifters_n = lifters.shape[0]
````

````python
football_players_mean = football_players.mean()
hockey_players_mean = hockey_players.mean()
lifters_mean = lifters.mean()

football_players_mean, hockey_players_mean, lifters_mean
````

````python
height = np.concatenate([football_players, hockey_players, lifters])
print(height)
````

````python
height_mean = height.mean()
height_mean
````

Значения $S_b$ и $S_{w}$:

````python
S2_b = football_players_n * (football_players_mean - height_mean) ** 2 + hockey_players_n * (hockey_players_mean - height_mean) ** 2 + lifters_n * (lifters_mean - height_mean) ** 2

S2_w = ((football_players - football_players_mean) ** 2).sum() + ((hockey_players - hockey_players_mean) ** 2).sum() + ((lifters - lifters_mean) ** 2).sum()

S2_b, S2_w
````

Проверим выполнение равенства $S^2 = S_b^2 + S_{w}^2:$

````python
round(S2_b + S2_w) == round(((height - height_mean) ** 2).sum())
````

Оценки дисперсий:

````python
k = 3
n = football_players_n + hockey_players_n + lifters_n

k1 = k - 1
k2 = n - k

sigma2_b = S2_b / k1
sigma2_w = S2_w / k2

sigma2_b, sigma2_w
````

Значение статистики $T$:

````python
T = sigma2_b / sigma2_w
T
````

Зафиксируем уровень значимости $\alpha = 0.05$. Для него найдём критическое значение $F_{crit}$:

````python
from scipy import stats
````

````python
alpha = 0.05

F_crit = stats.f.ppf(1 - alpha, k1, k2)
F_crit
````

````python
p = 1 - stats.f.cdf(T, k1, k2)
p
````

Видим, что $T > F_{crit}$, поэтому заключаем, что отличие среднего роста у представленных групп есть и является статистически значимым.

````python
from scipy.stats import f_oneway
f_oneway(football_players, hockey_players, lifters)
````
